```
import math
import numpy as np
def sigmoid(x):
  return 1/(1+math.exp(-x))

def sigmoid_np(x):
  return 1/(1+np.exp(-x))
```

```
x=1
y=0
w=tf.random.normal([1],0,1);
output=sigmoid(x*w)
output
```

```
output_np=sigmoid_np(x*w) # numpy함수로도 잘 작동
output_np
```

```
x=0
y=1

w=tf.random.normal([1],0,1)
b=tf.random.normal([1],0,1)

for i in range (1000):
  output=sigmoid(x*w+b)
  error=y-output
  w=w+x*0.1*error
  b=b+1*0.1*error

  if(i%100==99):
    print(i+1,error,output)
```

```
# AND 신경망
x=np.array([[1,1],[0,1],[1,0],[0,0]])
y=np.array([[1],[0],[0],[0]])
w=tf.random.normal([2],0,1);
b=tf.random.normal([1],0,1);
b_x=1

for i in range(2000):
  error_sum=0
  for j in range(4):
    output=sigmoid(np.sum(x[j]*w+b))
    error=y[j][0]-output
    w=w+x[j]*0.1*error # tensor에 저장된 numpy 배열에 대한 연산
    b=b+b_x*0.1*error
    error_sum+=error
  if i%200==199:
    print(i,error_sum)
  
```

```
print(w)
print(b)
```

```
for i in range(4):
  print('X:',x[i],'Y:',y[i],'Output:', sigmoid(np.sum(x[i]*w+b)))
```

```
for i in range(4):
  print('X:',x[i],'Y:',y[i],'Output:', sigmoid(np.sum(x[i]*w+b)))
```

```
# XOR 신경망
x=np.array([[1,1],[0,1],[1,0],[0,0]])
y=np.array([[0],[1],[1],[0]])
w=tf.random.normal([2],0,1);
b=tf.random.normal([1],0,1);
b_x=1

for i in range(2000):
  error_sum=0
  for j in range(4):
    output=sigmoid(np.sum(x[j]*w+b))
    error=y[j][0]-output
    w=w+x[j]*0.1*error # tensor에 저장된 numpy 배열에 대한 연산
    b=b+b_x*0.1*error
    error_sum+=error
  if i%200==199:
    print(i,error_sum)
```

```
x=np.array([[1,1],[0,1],[1,0],[0,0]])
y=np.array([[0],[1],[1],[0]])

model=tf.keras.Sequential([ #네트워크 배치
                           tf.keras.layers.Dense(units=2,activation='sigmoid',input_shape=(2,)), #layer1, dense type(fcl),units=뉴런수,input은 1차원 배열
                           tf.keras.layers.Dense(units=1,activation='sigmoid') # layer2, dense type
])

model.compile(optimizer=tf.keras.optimizers.SGD(lr=0.1),loss='mse'); # 모델 생성

model.summary() # 네트워크 구조 출력
```

```
history=model.fit(x,y,epochs=2000,batch_size=1) #네트워크 학습
```

```
model.predict(x)
```

```
for weight in model.weights:
  print(weight)
```
